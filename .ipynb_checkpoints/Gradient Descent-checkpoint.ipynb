Gradient Descent is perhaps the most intuitive of all optimization algorithms. Imagine you're standing on the side of a mountain and want to reach the bottom. You'd probably do something like this, Look around you and see which way points the most downwards Take a step in that direction, then repeat Well that's Gradient Descent! How does it work? So how do we frame Gradient Descent mathematically? As usual, we define our problem in terms of minimizing a function,

\begin{align*}
min_x\;f(x)
\end{align*}

We assume that $f$ is differentiable. That is, we can easily compute,

$$
  \nabla_x \, f(x) = \begin{pmatrix}
    \frac{d f(x)}{d x_1} \\
    \frac{d f(x)}{d x_2} \\
    \vdots \\
  \end{pmatrix}
$$


Given this, Gradient Descent is simply the following,

<strong>Input</strong>: initial iterate <mathjax>$x^{(0)}$</mathjax></p>
<ol>
<li>For <mathjax>$t = 0, 1, \ldots$</mathjax><ol>
<li>if converged, return <mathjax>$x^{(t)}$</mathjax></li>
<li>Compute the gradient of <mathjax>$f$</mathjax> at <mathjax>$x^{(t)}$</mathjax>, <mathjax>$g^{(t)}
   \triangleq \nabla f(x^{(t)})$</mathjax></li>
<li><mathjax>$x^{(t+1)} = x^{(t)} - \alpha^{(t)} g^{(t)}$</mathjax>
</div></li>
</ol>
</li>
</ol>

#### A Small Example

Let's look at Gradient Descent in action. We'll use the objective function $f(x) = x^4$, meaning that $\nabla_x f(x) = 4 x^3$. For a step size, we'll choose a constant step size $\alpha_t = 0.05$. Finally, we'll start at $x^{(0)} = 1$.

 Gradient Descent in action. The curved line is the $f(x)$, and the flat line is its linear approximation, $\hat{f}(y) = f(x) + \nabla_x f(x)^T (y-x)$, which is what Gradient Descent follows.
 This plot shows how quickly the objective function decreases as the number of iterations increases.
 This plot shows the actual iterates and the objective function evaluated at those points. More red indicates a higher iteration number.



<h1><a name="proof" href="#proof">Why does it work?</a></h1>
<p>Gradient Descent works, but it isn't guaranteed to find the optimal solution
to our problem (that is, <mathjax>$x^{*} = \arg\min_{x} f(x)$</mathjax>) without a few assumptions.
In particular,</p>
<ol>
<li><mathjax>$f$</mathjax> is convex and finite for all <mathjax>$x$</mathjax></li>
<li>a finite solution <mathjax>$x^{*}$</mathjax> exists</li>
<li><mathjax>$\nabla f(x)$</mathjax> is Lipschitz continuous with constant <mathjax>$L$</mathjax>. If <mathjax>$f$</mathjax> is twice
   differentiable, this means that the largest eigenvalue of the Hessian is
   bounded by <mathjax>$L$</mathjax> (<mathjax>$\nabla^2 f(x) \preceq LI$</mathjax>). But more directly, there must
   be an <mathjax>$L$</mathjax> such that,</li>
</ol>
<p><mathjax>$$
  || \nabla f(x) - \nabla f(y) ||_2 \le L || x - y ||_2 \qquad \forall x,y
$$</mathjax></p>
<p><strong>Assumptions</strong> So what do these assumptions give us?  Assumption 1 tells us
that <mathjax>$f$</mathjax> is lower bounded by an affine function,</p>
<p><mathjax>$$
  f(y) \ge f(x) + \nabla f(x)^T (y - x)  \qquad \forall x,y
$$</mathjax></p>
<p>Assumption 3 also tells us that <mathjax>$f$</mathjax> is upper bounded by a quadratic (this is
not obvious),</p>
<p><mathjax>$$
  f(y) \le f(x) + \nabla f(x)^T (y - x) + \frac{L}{2} || y - x ||_2^2
$$</mathjax></p>
<p><strong>Proof Outline</strong> Now let's dive into the proof. Our plan of attack is as
follows. First, we upper bound the error <mathjax>$f(x^{(t+1)}) - f(x^{*})$</mathjax> in terms of
<mathjax>$||x^{(t)} - x^{*}||_2^2$</mathjax> and <mathjax>$||x^{(t+1)} - x^{*}||_2^2$</mathjax>.  We then sum these upper
bounds across <mathjax>$t$</mathjax> to upper bound the sum of errors in terms of <mathjax>$||x^{(0)} -
x^{*}||_2^2$</mathjax>. Finally, we use the fact that <mathjax>$f(x^{(t)})$</mathjax> is decreasing in <mathjax>$t$</mathjax> to
take an average of that sum to bound <mathjax>$f(x^{(t+1)}) - f(x^{*})$</mathjax> in terms of <mathjax>$||x^{(0)}
- x^{*}||_2^2$</mathjax> and <mathjax>$t$</mathjax>.</p>
<p><strong>Step 1</strong>: upper bounding <mathjax>$f(x^{(t+1)}) - f(x^{*})$</mathjax>.  Let <mathjax>$x^{+} \triangleq
x^{(t+1)}$</mathjax>, <mathjax>$x \triangleq x^{(t)}$</mathjax>, and <mathjax>$\alpha \triangleq \alpha^{(t)}$</mathjax>.</p>
<p><mathjax>$$
\begin{align*}
  f(x^{+})
  \le &amp; f(x) + \nabla f(x)^T (x^{+} - x) + \frac{L}{2}||x^{+} - x||_2^2 &amp;&amp; \text{# Quadratic upper bound} \\
  = &amp; f(x) + \nabla f(x)^T (- \alpha \nabla f(x)) + \frac{L}{2}||- \alpha \nabla f(x)||_2^2 &amp;&amp; \text{# Definition of $x^{+}$} \\
  = &amp; f(x) - \alpha || \nabla f(x) ||_2^2 + \frac{\alpha^2 L}{2} ||\nabla f(x)||_2^2 \\
  = &amp; f(x) - \alpha\left( 1 - \frac{\alpha L}{2} \right) || \nabla f(x) ||_2^2  \\
  \le &amp; f(x) - \frac{\alpha}{2} || \nabla f(x) ||_2^2  &amp;&amp; \text{# Assuming $\frac{\alpha L}{2} \leq \frac{1}{2}$} \\
  \le &amp; f(x^{*}) + \nabla f(x)^T (x - x^{*}) - \frac{\alpha}{2} || \nabla f(x) ||_2^2  &amp;&amp; \text{# Linear lower bound on $f(x)$} \\
  = &amp; f(x^{*}) + \nabla f(x)^T (x - x^{*}) - \frac{\alpha}{2} || \nabla f(x) ||_2^2 \\
    &amp; \quad \pm \frac{1}{2 \alpha} \left( ||x||_2^2 + ||x^{*}||_2^2 + x^T x^{*} \right)\\
  = &amp; f(x^{*}) + \frac{1}{2 \alpha} \left(
      ||x - x^{*}||_2^2 - ||(x - \alpha \nabla f(x)) - x^{*}||_2^2
    \right) \\
  = &amp; f(x^{*}) + \frac{1}{2 \alpha} \left(
      ||x - x^{*}||_2^2 - ||x^{+} - x^{*}||_2^2
    \right) \\
\end{align*}
$$</mathjax></p>
<p><strong>Step 2</strong>: Upper bound <mathjax>$\sum_{t=1}^{T} f(x^{(t)}) - f(x^{*})$</mathjax> by summing across
all <mathjax>$t$</mathjax>. At this point we'll assume that <mathjax>$\alpha^{(t)}$</mathjax> is the same for all
<mathjax>$t$</mathjax>.</p>
<p><mathjax>$$
\begin{align*}
  f(x^{(t)}) - f(x^{*})
  &amp; \le \frac{1}{2 \alpha^{(t)}} \left(
    ||x^{(t)} - x^{*}||_2^2 - ||x^{(t+1)} - x^{*}||_2^2
  \right) \\
  \sum_{t=1}^{T} f(x^{(t)}) - f(x^{*})
  &amp; \le \frac{1}{2 \alpha} \sum_{t=1}^{T} \left(
    ||x^{(t)} - x^{*}||_2^2 - ||x^{(t+1)} - x^{*}||_2^2
  \right) \\
  &amp; = \frac{1}{2 \alpha} \left(
    ||x^{(0)} - x^{*}||_2^2 - ||x_1 - x^{*}||_2^2 + ||x_1 - x^{*}||_2^2 - ||x_2 - x^{*}||_2^2 + \ldots
  \right) \\
  &amp; = \frac{1}{2 \alpha} \left( ||x^{(0)} - x^{*}||_2^2 - ||x^{(t)} - x^{*}||_2^2
  \right) \\
  &amp; \le \frac{1}{2 \alpha} ||x^{(0)} - x^{*}||_2^2 \\
\end{align*}
$$</mathjax></p>
<p><strong>Step 3</strong>: Upper bound <mathjax>$f(x^{(t+1)}) - f(x^{*})$</mathjax> by using the fact that
<mathjax>$f(x^{(t+1)}) &lt; f(x^{(t)})$</mathjax></p>
<p><mathjax>$$
\begin{align*}
  f(x^{(T)}) - f(x^{*})
  &amp; \le \frac{1}{T} \sum_{t=1}^{T} ( f(x^{(t)}) - f(x^{*}) ) \\
  &amp; \le \frac{1}{2 \alpha T} ||x^{(0)} - x^{*}||_2^2
\end{align*}
$$</mathjax></p>

<p>Thus, we can conclude that if we want to reach an error tolerance <mathjax>$f(x^{T}) -
f(x^{*}) \le \epsilon$</mathjax>, we need <mathjax>$O(\frac{1}{\epsilon})$</mathjax> iterations.  In other
words, Gradient Descent has a "convergence rate" of <mathjax>$O(\frac{1}{T})$</mathjax>.</p>
<h1><a name="usage" href="#usage">When should I use it?</a></h1>
<p>Because it's so easy to implement, Gradient Descent should be the first thing
to try if you need to implement an optimization from scratch. So long as you
calculate the gradient right, it's practically impossible to make a mistake.
If you have access to an <a href="http://justindomke.wordpress.com/2009/02/17/automatic-differentiation-the-most-criminally-underused-tool-in-the-potential-machine-learning-toolbox/">automatic differentiation library</a> to do
the gradient computation for you, even better!  In addition, Gradient Descent
requires a minimal memory footprint, making it ideal for problems where <mathjax>$x$</mathjax> is
very high dimensional.</p>
<p>As we'll see in later posts, Gradient Descent trades memory for speed. The
number of iterations required to reach a desired accuracy is actually quite
large if you want accuracy on the order of <mathjax>$10^{-8}$</mathjax>, and there are algorithms
that are much faster if computation of the <a href="http://en.wikipedia.org/wiki/Hessian_matrix">Hessian</a> is feasible.
Even when considering the same memory requirements, there is another
gradient-based method with better convergence rates. </p>
<h1><a name="extensions" href="#extensions">Extensions</a></h1>
<p><strong>Step Size</strong> The proof above relies on a constant step size, but quicker
convergence can be obtained when using <a href="#line_search">Line Search</a>, wherein
<mathjax>$\alpha^{(t)}$</mathjax> is chosen to (approximately) find <mathjax>$\alpha^{(t)} = \arg\min_{\alpha}
f(x^{(t)} - \alpha \nabla f(x^{(t)}))$</mathjax>. Keep in mind that unless <mathjax>$0 \le t \le
\frac{1}{L}$</mathjax>, <em>Gradient Descent will not converge!</em></p>
<p><strong>Checking Convergence</strong> We have shown that the algorithm's error at iteration
<mathjax>$T$</mathjax> relies on <mathjax>$T$</mathjax> and the distance between <mathjax>$x^{(0)}$</mathjax> and <mathjax>$x^{*}$</mathjax>, the latter of
which is unknown.  How then can we check if we're "close enough"? A typical
choice is simply to stop after a fixed number of iterations, but another common
alternative is to quit when <mathjax> $||\nabla f(x^{(t)})||_2$ </mathjax> and <mathjax>$ \epsilon_{g}$ </mathjax> for a
chosen <mathjax>$\epsilon_{g}$</mathjax>.  The intuition for this comes from the assumption that
<mathjax>$f$</mathjax> is "strongly convex" with constant <mathjax>$m$</mathjax>, which then implies that <mathjax>$||x -
x^{*}||_2 \le \frac{2}{m}||\nabla f(x)||_2$</mathjax> (see <a href="http://www.stanford.edu/~boyd/cvxbook/">Convex
Optimization</a>, page 460, equation 9.10).</p>

<h1><a name="reference-impl" href="#reference-impl">Reference Implementation</a></h1>
<p>Here's a quick implementation of gradient descent,</p>

```python
%matplotlib inline
import numpy as np
import pylab as pl
from numpy import newaxis, r_, c_, mat
from numpy.linalg import *

def gradient_descent(gradient, x0, alpha, n_iterations=100):
    xs = [x0]
    for t in range(n_iterations):
        x = xs[-1]
        g = gradient(x)
        x_plus = x - alpha(t) * g
        xs.append(x_plus)
    return xs

# This generates the plots that appear above
if __name__ == '__main__':
    func = lambda x: x ** 4  # the function to minimize
    gradient = lambda x: 4 * x **3  # its gradient
    step_size = 0.05
    x0 = 1.0
    n_iterations = 10

  # run gradient descent
    iterates = gradient_descent(gradient, x0, lambda x: step_size, n_iterations=n_iterations)
    i=1
  ### PLOTTING ###
    for t in range(n_iterations):
        x = iterates[t]
        x_plus = iterates[t+1]

        f = func
        g = gradient
        f_hat = lambda y: f(x) + g(x) * (y - x)

        x_min = (0-f(x))/g(x) + x
        x_max = (1.1-f(x))/g(x) + x

        fig = pl.figure()
        fig.add_subplot(2, 5, i)
        
        pl.plot(np.linspace(0, 1.1, 100), func(np.linspace(0, 1.1, 100)), alpha=0.2)
        pl.xlim([0, 1.1])
        pl.ylim([0, 1.1])
        pl.xlabel('x')
        pl.ylabel('f(x)')

        pl.plot([x_min, x_max], [f_hat(x_min), f_hat(x_max)], '--', alpha=0.2)
        pl.scatter([x, x_plus], [f(x), f(x_plus)], c=[0.8, 0.2])

        i = i + 1
        pl.show()
```
